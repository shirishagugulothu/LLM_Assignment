```python
#import required module
from huggingface_hub import InferenceClient
from getpass import getpass
import textwrap # textwrap wraps the output in multiple lines

# Hugging Face API token
HF_TOKEN = getpass("Enter your Hugging Face API token: ")

# Load model client
# create an inference client
# InferenceClient - Python class from the huggingface_hub
client = InferenceClient(
    model="mistralai/Mistral-7B-Instruct-v0.2",
    token=HF_TOKEN
)

# Ask the user a question or a prompt
user_input = input("Enter your question: ")

# Chat-style message
messages = [
    {"role": "user", "content": user_input}
]

# Get response
response = client.chat_completion(
    messages=messages,
    max_tokens=200,

)

# diplay structured result
print("\nAnswer:\n")
print(textwrap.fill(response.choices[0].message.content.strip(), width=80))
```

```python
# Import the required classes from the transformers library
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch  # For tensor operations

# Step 1: Specify the model you want to use (FLAN-T5 base)
model_name = "google/flan-t5-base"

# Step 2: Load the tokenizer for the specified model
# The tokenizer converts input text into tokens (numerical format) the model understands
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Step 3: Load the pre-trained model for sequence-to-sequence tasks
# FLAN-T5 is good for tasks like Q&A, summarization, translation, etc.
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Step 4: Define a function that generates output from a prompt
def generate_output(prompt, max_tokens=100):
    # Step 4.1: Tokenize the prompt text into input tensors
    inputs = tokenizer(prompt, return_tensors="pt", padding=True)

    # Step 4.2: Generate output from the model based on the input
    # `max_new_tokens` controls how long the generated output can be
    outputs = model.generate(**inputs, max_new_tokens=max_tokens)

    # Step 4.3: Decode the output tokens back into human-readable text
    result = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Step 4.4: Return the final result
    return result
```

```python
# Task: TEXT GENERATION

# Zero-shot
prompt1 = "Write a short story about a robot learning to love."
print("Zero-shot:\n", generate_output(prompt1))
# Comment: Model gives a short story, basic structure is present.

# Few-shot
prompt2 = """Write a short story:
Story 1: A dog learned to dance with its owner and became famous.
Story 2: A tree befriends a lonely boy and helps him grow strong.
Story 3: A robot learns to love after spending time with a child."""
print("Few-shot:\n", generate_output(prompt2))
# Comment: More creative and structured output due to examples.

# Chain-of-thought
prompt3 = """Write a short story about a robot learning to love.
Step 1: Introduce the robot's function and lonely life.
Step 2: Show how it meets a human.
Step 3: Explain how emotional connection develops.
Step 4: Show how the robot changes.
"""
print("Chain-of-thought:\n", generate_output(prompt3))
# Comment: Better flow and emotional depth with guided steps.
```

```python
# Task: SUMMERIZATION

text = """Artificial intelligence is changing how healthcare is delivered.
It helps in faster diagnosis, personalized treatment, and administrative automation.
Doctors can now spend more time with patients."""

# Zero-shot
prompt = "Summarize: " + text
print("Zero-shot Summary:\n", generate_output(prompt))
# Comment: Provides basic summary, captures key idea.

# Few-shot
prompt = """Summarize the following:
Text: Students who read often develop stronger vocabularies and better comprehension.
Summary: Reading improves vocabulary and understanding.

Text: Artificial intelligence is changing how healthcare is delivered.
It helps in faster diagnosis, personalized treatment, and administrative automation.
Doctors can now spend more time with patients.
Summary:"""
print("Few-shot Summary:\n", generate_output(prompt))
# Comment: More concise and relevant summary.

# Chain-of-thought
prompt = """Summarize the text step-by-step:
Text: Artificial intelligence is changing how healthcare is delivered.
It helps in faster diagnosis, personalized treatment, and administrative automation.
Doctors can now spend more time with patients.

Step 1: Identify main idea.
Step 2: Extract key contributions.
Step 3: Write a concise summary."""
print("Chain-of-thought Summary:\n", generate_output(prompt))
# Comment: Output is more refined with guided thinking.
```

```python
# TASK: QUESTION AND ANSWER

# Zero-shot
prompt = "Question: What is the capital of France? Answer:"
print("Zero-shot QA:\n", generate_output(prompt))
# Comment: Correct answer, straightforward.

# Few-shot
prompt = """Question: What is the capital of Germany? Answer: Berlin
Question: What is the capital of Italy? Answer: Rome
Question: What is the capital of France? Answer:"""
print("Few-shot QA:\n", generate_output(prompt))
# Comment: Confident and consistent answers.

# Chain-of-thought
prompt = """Question: What is the capital of France?
Let's think step-by-step:
1. France is a country in Europe.
2. The largest and most well-known city is Paris.
3. Paris is the capital city.
Answer:"""
print("Chain-of-thought QA:\n", generate_output(prompt))
# Comment: Explains reasoning clearly, useful in complex questions.
```

```python
# TASK: TRANSLATION

# Zero-shot
prompt = "Translate 'Good morning, how are you?' to French."
print("Zero-shot Translation:\n", generate_output(prompt))
# Comment: Works accurately without context.

# Few-shot
prompt = """Translate the following:
English: Hello -> French: Bonjour
English: Thank you -> French: Merci
English: Good morning, how are you? -> French:"""
print("Few-shot Translation:\n", generate_output(prompt))
# Comment: Better fluency and word selection.

# Chain-of-thought
prompt = """In a box there are 5 green apples and 4 red apples. Total how many apples are there in the box?"""
print("Chain-of-thought Translation:\n", generate_output(prompt))
# Comment: Explicit logical steps help improve language understanding.
```
